## 10-2. 데이터 단순 합치기
* append : 행 합치기만 가능
* concat : 행 합치기, 열 합치기(axis=1) 모두 가능

```python
# 1. append (행 합치기)
df_copy = df.copy()
df_copy.head()
```

```python
df_append = df.append(df_copy)  # ignore_index=True
df_append
```

```python
# 2. concat (행 합치기)
df_concat = pd.concat([df, df_copy], ignore_index=True)
df_concat
```

```python
# 2. concat (열 합치기)
df2 = pd.read_csv('http://bit.ly/ds-korean-idol-2')
df2
```

```python
# 행 인덱스 라벨을 기준으로 붙임
df_concat = pd.concat([df, df2], axis=1)
df_concat
```

```python
# 행 인덱스 라벨 1 삭제
df.drop(1, inplace=True)
df
```

```python
# 행 인덱스 라벨 3, 5번 삭제
df2.drop([3, 5], inplace=True)
df2
```

```python
df_concat = pd.concat([df, df2], axis=1)
df_concat
```

```python
# concat은 인덱스 라벨을 기준으로 데이터를 붙이므로 인덱스 라벨이 달라지면 데이터 오류가 발생할 수 있음
df.reset_index(drop=True, inplace=True)
df
```

## 10-3. 데이터프레임 병합하기
* merge : 특정 키를 기준으로 데이터프레임 병합하기 -> df와 df2의 이름을 기준으로 병합하기
* 데이터베이스의 Join과 유사함
* pd.merge(left, right, on, how)
    * left와 right는 병합할 두 DataFrame을 대입합니다.
    * on 에는 병합의 기준이 되는 column을 넣어 줍니다.
    * how 에는 'left', 'right', 'inner', 'outer' 라는 4가지의 병합 방식중 한가지를 택합니다.    

### 1) left, right 방식

```python
df
```

```python
df2
```

```python
# how='left' 옵션을 주면, left DataFrame을 기준으로 병합이 됨 (행 갯수 14개)
# left outer join과 동일
pd.merge(df, df2, on='이름', how='left')
```

```python
# how='right' 옵션을 주면 right DataFrame을 기준으로 병합이 됨 (행 갯수 13개)
# right outer join과 동일
pd.merge(df, df2, on='이름', how='right')
```

### 2) inner, outer 방식
* inner 방식은 두 DataFrame에 모두 키 값이 존재하는 경우만 병합합니다.
* outer 방식은 하나의 DataFrame에 키 값이 존재하는 경우 모두 병합합니다.
* outer 방식에서는 없는 값은 NaN으로 대입됩니다.

```python
# 데이터의 갯수 12개
pd.merge(df, df2, on='이름', how='inner')
```

```python
# 데이터의 갯수 15개
# full outer join과 동일
# left를 먼저 보여주고, right에서 빠진 데이터 추가
pd.merge(df, df2, on='이름', how='outer')
```

```python
# column명은 다르지만, 동일한 성질의 데이터인 경우 병합하기
df2
```

```python
df2.columns = ['성함', '연봉', '가족수']
df2
```

```python
# 동일한 데이터이지만 컬럼명이 다를 경우 left_on, right_on 사용하기
pd.merge(df, df2, left_on='이름', right_on='성함', how='outer')
```
