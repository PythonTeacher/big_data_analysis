# 3. Scatter Plot
* 좌표상의 점들을 표시함으로써 두 변수 간의 관계를 나타내는 그래프 
* 두 연속형 변수의 상관관계를 알아내는 데 유용함

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#np.random.seed(2021)
x = np.random.rand(50)  # shape에 맞게 0이상 1미만의 난수 생성 -> [0, 1)
y = np.random.rand(50)
plt.scatter(x, y)  # plt.plot(x, y, 'o')
plt.show()
```

```python
# scatter plot 2개 그리기
plt.figure(figsize=(10, 10))
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)

x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)

plt.show()
```

```python
# Marker 지정하기
plt.scatter(x, y, marker='*')   # marker
plt.show()
```

```python
# Color 지정하기
plt.scatter(x, y, c='r')   # color
plt.show()
```

```python
# 점마다 다른 Color 지정하기
x = [1, 2, 3]
y = [1, 2, 3]
colors=['r', 'g', 'b']
plt.scatter(x, y, c=colors)   # color
plt.show()
```

```python
# ColorMap(색상지도) 사용하기
# https://matplotlib.org/stable/tutorials/colors/colormaps.html

# 데이터를 기반으로 색상 표현하기
# 예) 파란색과 빨간 평균 기온에 따라 추운 지역은 파란색으로, 더운 지역은 빨간색으로 표시하기
# 모든 지역은 파란색과 빨간색 사이의 그라데이션에 포함된 색상으로 표현됨 -> 이 색상표를 colormap이라고 함
plt.colormaps()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, c=x, cmap='Blues')  # cmap=plt.cm.Blues
plt.colorbar()
plt.show()
```

```python
# Marker Size 변경하기
size = np.random.rand(50) * 100  # 0.1 * 100 -> 10
plt.scatter(x, y, c=x, cmap='Blues', s=size)
plt.show()
```

```python
# 투명도 (alpha값은 투명도를 나타내며 0~1 사이의 값을 지정해 줄 수 있으며, 0에 가까울수록 투명한 값을 가짐)
plt.scatter(x, y, c=x, cmap='Blues', s=size, alpha=0.5)
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 5))

im1 = axes[0].scatter(x, y, c=y, cmap=plt.cm.viridis, s=size, alpha=0.1)
im2 = axes[1].scatter(x, y, c=y, cmap=plt.cm.viridis, s=size, alpha=0.5)
im3 = axes[2].scatter(x, y, c=y, cmap=plt.cm.viridis, s=size, alpha=1.0)

axes[0].set_title('Alpha = 0.1')
axes[1].set_title('Alpha = 0.5')
axes[2].set_title('Alpha = 1.0')

fig.colorbar(im1, ax=axes[0])
fig.colorbar(im2, ax=axes[1])
fig.colorbar(im3, ax=axes[2])

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(131)
plt.scatter(x, y, c=y, cmap='viridis', s=size, alpha=0.1)
plt.title('alpha=0.1')
plt.colorbar()

plt.subplot(132)
plt.title('alpha=0.5')
plt.scatter(x, y, c=y, cmap='viridis', s=size, alpha=0.5)
plt.colorbar()

plt.subplot(133)
plt.title('alpha=1.0')
plt.scatter(x, y, c=y, cmap='viridis', s=size, alpha=1.0)
plt.colorbar()

plt.tight_layout()
plt.show()
```
