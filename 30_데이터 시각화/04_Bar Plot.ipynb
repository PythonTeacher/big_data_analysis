# 4. Bar Plot
* 막대 그래프로 여러 개의 값을 서로 비교하는 데 유용함
* matplotlib.pyplot의 bar() 함수를 이용해서 막대 그래프를 표현할 수 있음
* x축에는 범주형 변수를, y축에는 연속형 변수를 넣음

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 년도별 출생자수 비교
years = ['2018', '2019', '2020']
number = [100, 90, 80]

plt.bar(years, number, width=0.5, color='pink')
plt.title('년도별 출생자수', size=20)
plt.xlabel('년도')
plt.ylabel('출생자수')
plt.show()
```

```python
# 수평 막대그래프 그리기
plt.barh(years, number, height=0.5)
plt.title('년도별 출생자수', size=20)
plt.xlabel('출생자수')
plt.ylabel('년도')
plt.show()
```

```python
# 년도, 성별로 출생자수 비교하기
years = ['2018', '2019', '2020']
male = [100, 90, 80]
female = [90, 80, 70]
x = np.arange(3)

plt.bar(x, male, width=0.6)
plt.bar(x, female, width=0.6)

plt.xticks(x, years)
plt.legend(['male', 'female'])

plt.title('년도별 출생자수', size=20)
plt.xlabel('년도')
plt.ylabel('출생자수')
plt.show()
```

```python
years = ['2018', '2019', '2020']
male = [100, 90, 80]
female = [90, 80, 70]

x = np.arange(3)  # 0~2
bar_width = 0.3

plt.bar(x - bar_width/2, male, width=bar_width)
plt.bar(x + bar_width/2, female, width=bar_width)

plt.xticks(x, years)
plt.legend(['male', 'female'])

plt.title('년도별 출생자수', size=20)
plt.xlabel('년도')
plt.ylabel('출생자수')
plt.show()
```

```python
years = ['2018', '2019', '2020']
male = [100, 90, 80]
female = [90, 80, 70]

x = np.arange(len(years))  # 0~2
bar_width = 0.3

plt.barh(x - bar_width/2, male, height=bar_width)
plt.barh(x + bar_width/2, female, height=bar_width)

plt.yticks(x, years)
plt.legend(['male', 'female'])

plt.title('년도별 출생자수', size=20)
plt.xlabel('출생자수')
plt.ylabel('년도')
plt.show()
```
