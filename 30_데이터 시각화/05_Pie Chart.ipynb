# 5. Pie Chart
* 카테고리별 또는 그룹별로 집계를 하고 이를 백분율로 표시할 때 유용
* 데이터의 비율을 직관적으로 알 수 있음

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = [1, 2, 3, 4]

plt.pie(data)
plt.show() 
```

```python
labels = ['피자', '떡볶이', '자장면', '치킨']
plt.pie(data, labels=labels)
plt.show()
```

```python
# 자동으로 퍼센티지를 텍스트로 표시하기
labels = ['피자', '떡볶이', '자장면', '치킨']
plt.pie(data, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
labels = ['피자', '떡볶이', '자장면', '치킨']
explode = [0.2, 0, 0, 0]

plt.pie(data, labels=labels, autopct='%.1f%%', explode=explode, shadow=True)
plt.show()
```

```python
labels = ['피자', '떡볶이', '자장면', '치킨']
explode = [0.2, 0, 0, 0]
colors = ['r', 'g', 'b', 'pink']
plt.pie(data, labels=labels, autopct='%.1f%%', explode=explode, shadow=True, colors=colors)
plt.show()
```
