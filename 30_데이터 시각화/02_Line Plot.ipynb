# 2. Line Plot
* 시간에 따른 데이터의 변화를 시각화하는데 유용함
* 예) 아파트 가격의 변동, 지난 두달간 몸무게의 변화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 2-1. 그래프 그리기

```python
# 두 점 (1, 3), (8, 10)을 연결하는 선 그래프 그리기
x = [1, 8]
y = [3, 10]
plt.plot(x, y)  # 점을 연결하는 선 그래프를 그림
plt.show()
```

```python
# 선을 제거하고 두 점만 그리기 (o: rings)
# marker 종류 여러개
plt.plot(x, y, 'o')
plt.show()
```

```python
# (1, 2) ~ (10, 20)
x = np.arange(1, 11)
y = x * 2
plt.plot(x, y, 'o')
plt.show()
```

```python
# 점과 선 둘다 그리기
plt.plot(x, y, marker='o')
plt.show()
```

## 2-2. Marker, Line, Color
* Marker : https://matplotlib.org/stable/api/markers_api.html?highlight=marker
* Line : https://matplotlib.org/3.0.3/gallery/lines_bars_and_markers/line_styles_reference.html
* Color : https://matplotlib.org/stable/gallery/color/named_colors.html
* Format String(fmt) : Marker:Line:Color

```python
# Marker
plt.plot(x, y, marker='*')
plt.show()
```

```python
# Marker size
plt.plot(x, y, marker='*', ms=10)
plt.show()
```

```python
# Line Style
# - (solid line)
# : (dotted line)
# -- (dashed line)
# -. (dashdot line)
plt.plot(x, y, linestyle='solid')
plt.plot(y, x, ls=':')
plt.show()
```

```python
# Color
plt.plot(x, y, color='r')
plt.plot(y, x, color='b', linewidth=5)  # lw
plt.show()
```

```python
# Format String
plt.plot(x, y, 'o:r')
plt.plot(y, x, '*--b')
plt.show()
```

```python
# 그래프 분리하기
# figure()는 새로운 그래프 Canvas를 생성합니다.
plt.plot(x, y, 'o:r')

plt.figure()
plt.plot(y, x, '*--b')
plt.show()
```

```python
# 그래프 사이즈 변경하기 (단위 : inch)
plt.figure(figsize=(5,5))
plt.plot(x, y, 'o:r')

plt.figure(figsize=(10, 5))
plt.plot(y, x, '*--b')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 5)
y = x * 2
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o:r')
plt.plot(y, x, '*--b')
plt.show()
```

## 1-3. Title, Label, Tick, Legend

```python
# 타이틀
plt.title('Title')
plt.plot(x, y)
plt.show()
```

```python
# 타이틀 사이즈 변경하기
plt.plot(x, y)
plt.title('Title', fontsize=20)
plt.show()
```

```python
# Label 설정
plt.plot(x, y)
plt.title('Title', fontsize=20)
plt.xlabel('X', fontsize=15)
plt.ylabel('Y', fontsize=15)
plt.show()
```

```python
# Tick 설정
# Tick은 X, Y축에 위치한 눈금을 의미함
plt.plot(x, y)

# X tick, Y tick 설정
plt.xticks(rotation=90)
plt.yticks(rotation=30)

plt.show()
```

```python
# legend(범례) 설정
plt.plot(x, y)
plt.plot(y, x)

plt.legend(['(x, y)', '(y, x)'], fontsize=15)
plt.show()
```

```python
# legend(범례) 설정 2
plt.plot(x, y, label='(x, y)')
plt.plot(y, x, label='(y, x)')

plt.legend(fontsize=15)
plt.show()
```

```python
height = [163, 167, 170, 175]
weight = [55, 58, 63, 67]
plt.plot(height, weight)
plt.title('키에 따른 몸무게 그래프')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.legend(['x=키, y=몸무게'])
plt.show()
```

## 1-5. Subplot
* 하나의 figure에 여러개 plot 그리기
* subplot()
* add_subplot()
* subplots()

```python
# 하나의 figure에 여러개 plot 그리기
# 1. subplot(row, column, index)
plt.subplot(1, 2, 1)
plt.plot(x, y)

plt.subplot(1, 2, 2)
plt.plot(y, x)

plt.show()
```

```python
# 콤마를 생략하고 row, column, index로 작성가능
# 211 -> row: 2, col: 1, index: 1
plt.subplot(211)
plt.plot(x, y)
plt.title('(x, y)')

plt.subplot(212)
plt.plot(y, x)
plt.title('(y, x)')

plt.suptitle('Title', size=20)  # 전체 타이틀 붙이기
plt.tight_layout() # 자동 레이아웃 설정
plt.show()
```

```python
# 2. add_subplot()
# Axes와 Axis의 차이
# https://stackoverflow.com/questions/5575451/difference-between-axes-and-axis-in-matplotlib
# https://towardsdatascience.com/matplotlib-seaborn-pandas-an-ideal-amalgamation-for-statistical-data-visualisation-f619c8e8baa3
fig = plt.figure()
axes1 = fig.add_subplot(2, 1, 1)
axes2 = fig.add_subplot(2, 1, 2)

axes1.plot(x, y)
axes1.set_title('(x, y)')

axes2.plot(y, x)
axes2.set_title('(y, x)')

plt.tight_layout()
plt.show()
```

```python
# 3. subplots()
fig, axes = plt.subplots(2, 1)  # nrows=2, ncols=1

axes[0].plot(x, y)
axes[1].plot(y, x)

axes[0].set_title('(x, y)')
axes[1].set_title('(y, x)')

plt.tight_layout()
plt.show()
```
